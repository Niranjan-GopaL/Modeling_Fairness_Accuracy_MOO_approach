```python
import pandas as pd

# Load CSV 
df = pd.read_csv("pareto_results.csv")

# Clean model names (strip whitespace)
df['model'] = df['model'].str.strip()
df['optimization_strat'] = df['optimization_strat'].str.strip()

# Define group keys
df['group'] = df['optimization_strat'].str.lower() + "_" + df['model'].str.lower()

# Split into 6 groups
grouped_dfs = {}
for group in df['group'].unique():
    grouped_dfs[group] = df[df['group'] == group].reset_index(drop=True)
    print(f"\n=== {group.upper()} ===")
    display(grouped_dfs[group])
```

```python
import matplotlib.pyplot as plt

# Assign colors manually
color_map = {
    'botorch_nn': 'red',
    'optuna_nn': 'blue',
    'botorch_rf': 'green',
    'optuna_rf': 'orange',
    'botorch_lr': 'purple',
    'optuna_lr': 'brown'
}

plt.figure(figsize=(14, 12))

for group, group_df in grouped_dfs.items():
    color = color_map.get(group, 'gray')
    x = group_df['accuracy']
    y = group_df['dpd']
    plt.scatter(x, y, label=group.upper(), color=color)
    for i, (xi, yi) in enumerate(zip(x, y)):
        plt.annotate(str(i), (xi, yi), textcoords="offset points", xytext=(5, 5), fontsize=8)

plt.xlabel("Accuracy")
plt.ylabel("Demographic Parity Difference (DPD, lower is better)")
plt.title("All Optimization Groups with Point Indices")
plt.legend()
plt.grid(True)
plt.show()
```

```python
combined = []
for group, group_df in grouped_dfs.items():
    for i, row in group_df.iterrows():
        combined.append({
            'accuracy': row['accuracy'],
            'dpd': row['dpd'],
            'group': group,
            'index': i,
            'color': color_map[group],
            'params': row['model_parameter_string']
        })

# Determine dominance
def is_dominated(p1, p2):
    return (p2['accuracy'] >= p1['accuracy'] and p2['dpd'] <= p1['dpd']) and (
        p2['accuracy'] > p1['accuracy'] or p2['dpd'] < p1['dpd'])

pareto_front = []
for p1 in combined:
    if not any(is_dominated(p1, p2) for p2 in combined if p2 != p1):
        pareto_front.append(p1)

# Prepare plot
plt.figure(figsize=(10, 8))

# Plot all points (non-Pareto as white with gray edge)
for p in combined:
    if p in pareto_front:
        plt.scatter(p['accuracy'], p['dpd'], color=p['color'])
        plt.annotate(f"{p['index']}", (p['accuracy'], p['dpd']), textcoords="offset points", xytext=(5, 5), fontsize=8)
    else:
        plt.scatter(p['accuracy'], p['dpd'], color='white', edgecolor='gray')

# Sort Pareto points for line plotting
pareto_sorted = sorted(pareto_front, key=lambda p: (p['accuracy'], -p['dpd']))
pareto_x = [p['accuracy'] for p in pareto_sorted]
pareto_y = [p['dpd'] for p in pareto_sorted]
plt.plot(pareto_x, pareto_y, color='yellow', linewidth=2, linestyle='-', label='Pareto Front')


plt.xlabel("Accuracy")
plt.ylabel("Demographic Parity Difference (DPD, lower is better)")
plt.title("Combined Pareto Front (Only Pareto Points Colored and Labeled)")
plt.grid(True)
plt.legend()
plt.show()

```

```python
# Store Pareto points in a DataFrame and save to CSV
pareto_df = pd.DataFrame(pareto_front)
pareto_df.to_csv("final_pareto.csv", index=False)

# Display the Pareto points with model parameters
print("=== Pareto Points with Parameters ===\n")
for p in pareto_front:
    print(f"[{p['group'].upper()}] Index: {p['index']}, Accuracy: {p['accuracy']:.4f}, DPD: {p['dpd']:.4f}")
    print(f"    Params: {p['params']}\n")
```

```python
import pandas as pd
import re

input_csv = 'final_pareto_of_combined_ensamble.csv'
output_csv = 'cleaned_pareto_models.csv'

df = pd.read_csv(input_csv)

def clean_params(params_str):
    if isinstance(params_str, str) and params_str.startswith('{') and params_str.endswith('}'):
        # Replace commas followed by spaces inside the parameter string
        return params_str.replace(', ', ' ')
    return params_str

df['params'] = df['params'].apply(clean_params)

df = df.sort_values(by='accuracy', ascending=False)

df.to_csv(output_csv, index=False)

print(f"CSV file cleaned and sorted by accuracy (highest first). Output saved to {output_csv}")
```
